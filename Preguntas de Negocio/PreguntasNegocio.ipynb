```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer datos desde el archivo CSV
df = pd.read_csv('/content/bank-full.csv', sep=';')

# Calcular tasa de conversin
conversion_data = df.groupby('contact')['y'].value_counts(normalize=True).unstack().fillna(0)
conversion_data['conversion_rate'] = conversion_data['yes']  # Tasa de conversin
conversion_data = conversion_data.reset_index()

# Grfico de lneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=conversion_data, x='contact', y='conversion_rate', marker='o')
plt.title('Tasa de Conversin por Tipo de Contacto')
plt.xlabel('Tipo de Contacto')
plt.ylabel('Tasa de Conversin')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Grfico de violn
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='contact', y='duration', hue='y', split=True)
plt.title('Duracin de las Llamadas por Tipo de Contacto y Resultado')
plt.xlabel('Tipo de Contacto')
plt.ylabel('Duracin de la Llamada (segundos)')
plt.grid(True)
plt.show()
```
